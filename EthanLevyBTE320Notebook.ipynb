```python
ales= float(input("Enter the sales value ($):"))
COGS= sales *0.40
GrossProfit= sales - COGS
print(f"Sales: {sales}$")
print(f"COGS: {COGS}$")
print(f"GrossProfit: {GrossProfit}$")
Taxrate= float(input("Enter the tax rate (%):"))
Taxes= GrossProfit * Taxrate
NetIncome= GrossProfit - Taxes
print(f"Taxes: {Taxes}$")
print(f"NetIncome: {NetIncome}$")
```
