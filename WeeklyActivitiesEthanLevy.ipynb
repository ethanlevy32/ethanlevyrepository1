Week 3 Activity

Design a simple numerical calculator. The user provides three inputs: two numbers and the operator. The program will print the output on the screen.

```python
num1 = float(input("Enter a number:"))
num2 = float(input("Enter another number:"))
operation = input("Enter an operation (+, -, *, /):")

if operation == "+":
    print(num1 + num2)
elif operation == "-":
    print(num1 - num2)
elif operation == "*":
    print(num1 * num2)
elif operation == "/":
    print(num1 / num2)
else: print("An error occurred during calculation. You can't divide by 0.")
```

Parking garage charges \$5 plus $2.50 each hour. Write a program that shows the number of hours parked and the corresponding fee side by side from 1 to 8 hours.
Also do a minimum and maximum fee of 10 and 20 respectively.

```python
for hour in range(1,9,1):
  fee= 5 +2.5*hour
  if fee<10:
    fee=10.0
  elif fee>20:
    fee=20.0
  print(hour, fee)
```

At a movie theater, tickets are \$10. Without advertisng 20 people attend. More people come depending on marketing: Additional= 2 x advertising dollars ^1/2. Theres also a fixed cost to run the theater of \$200.

```python
import math

ticket= 10
init_attendees= 20
fixed_cost = 200

print("Ad Profit")

for Ad in range(0,201,25):
  revenue= (init_attendees + 2 * round(math.sqrt(Ad)))*10
  costs= fixed_cost + Ad
  profit= revenue-costs
  print(Ad, profit)
```

Week 6 Activity

Rock paper scissors activity

```python
import random #

play = input("Do you want to play (y/n)")


while play == "y":
  player_move= (input("Rock, paper, or scissors?"))
  computer_move= (random.choice(['rock', 'paper', 'scissors']))

  if player_move == computer_move:
    print("Draw")
  elif player_move == "rock" and computer_move == "scissors":
    print("You win!")
  elif player_move == "rock" and computer_move == "paper":
    print("You lose.")
  elif player_move == "paper" and computer_move == "rock":
    print("You win!")
  elif player_move == "paper" and computer_move == "scissors":
    print("You lose.")
  elif player_move == "scissors" and computer_move == "paper":
    print("You win!")
  elif player_move == "scissors" and computer_move == "rock":
    print("You lose.")

  play = input("Do you want to play again? (y/n)")
```
